```python
import pandas as pd
import numpy as np
import pylab
import matplotlib
import matplotlib.pyplot as plt
import mplfinance as mpf
import seaborn as sns
import statsmodels.api as sm
from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.tsa.stattools import adfuller
import scipy.stats as stats
%matplotlib inline
```

```python
#Input
Symbol = "EURUSD"
year = "2017"
start_month = 1
end_month = 2
```

```python
path = "C:/Users/__/Python file/Data/01_"+str(Symbol)+"_Tick_28%/"+str(year)+"_"+str(start_month)+"_"+str(end_month)+'.csv'

df = pd.read_csv(path,index_col=0,header=0) #Reading data in file
df['timestamp'] = pd.to_datetime(df['timestamp'])
df = df.set_index(['timestamp'])
df
```

```python
Holiday = ['2017-01-07','2017-01-08','2017-01-14','2017-01-15','2017-01-21','2017-01-22','2017-01-28','2017-01-29',
            '2017-02-04','2017-02-05','2017-02-11','2017-02-12','2017-02-18','2017-02-19','2017-02-25','2017-02-26']
```

```python
Holiday_filter = []
for H in Holiday:
    Holiday_filter.append(df[H].index)
    
#H = df['2017-02-25'].index
#Holiday_filter

for D in range(0,len(Holiday_filter)):
    df = df.drop(index=Holiday_filter[D])
#df
```

```python
df['Ret_close'] = df.close.pct_change(1)*100

Candle_Color = []
Candle_Color.append("NaN")
for i in range(0,len(df)-1):
    if df.close[i+1]>df.close[i]:
        Candle_Color.append("Green")
    if df.close[i+1]==df.close[i]:
        Candle_Color.append("Yellow")
    if df.close[i+1]<df.close[i]:
        Candle_Color.append("Red")
        
df['Candle_Color'] = Candle_Color
```

```python
df = df.sort_values('timestamp',ascending=False)
```

```python
Accum_Ret = []
Accum_Return = 0

for i in range(0,len(df)):
    
    Accum_Return = Accum_Return+df['Ret_close'][i]
    Accum_Ret.append(Accum_Return)
    #print(Accum_Return)

df['Accum_Ret'] = Accum_Ret

df
```

```python
result = adfuller(df.Ret_close.dropna())
adf_statistic = result[0]
p_value = result[1]
print("ADF Statistic:", adf_statistic)
print("p-value:", p_value)
```

```python
acf,confint = sm.tsa.acf(df.Ret_close.dropna(),nlags=100,alpha=0.05)
plt.figure(figsize=(15,6))
plt.stem(acf[1:len(df)])

plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.show()
```

```python
pacf = sm.tsa.pacf(df.Ret_close.dropna(), nlags=100)
plt.figure(figsize=(15,6))
plt.stem(pacf[1:len(df)])

plt.xlabel('Lag')
plt.ylabel('Partial Autocorrelation')
plt.show()
```

```python
Data = df.Ret_close.dropna()
null_value = 0
t_statistic , p_value = stats.ttest_1samp(Data,null_value)
confidence_interval = stats.t.interval(0.95, len(Data)-1, loc=Data.mean(), scale=stats.sem(Data))

print("t_statistic ",t_statistic)
print("p_value ",p_value)
print("95% Confidence Interval:", confidence_interval)
```

```python
df['low_confidence_0'] = confidence_interval[0]
df['low_confidence_1'] = confidence_interval[1]
```

```python
significant_move = []
significant_move.append(0)
for i in range(1,len(df)):
    if (df["Ret_close"][i]<confidence_interval[0]) or (df["Ret_close"][i]>confidence_interval[1]):
        significant_move.append(df["Ret_close"][i])
    else:
        significant_move.append(0)
```

```python
df['significant_move'] = significant_move
df
```

```python
Accum_Significant = []
Sum_Significant = df.significant_move.sum()

for i in range(0,len(df)):
    
    Sum_Significant = Sum_Significant-df['significant_move'][i]
    Accum_Significant.append(Sum_Significant)


df['Accum_Significant'] = Accum_Significant
df
```
